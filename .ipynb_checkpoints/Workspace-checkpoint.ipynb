# Predicting heart disease using machine learning

## 1. Problem Definition 
    
> Given some features/data about a patient we would like to predict whether someone has a heart disease or not.

## 2. Data 

> This data is taken from University of California, Irvine which is available on https://archive.ics.uci.edu/ml/datasets/heart+disease

> It is also available on kaggle in https://www.kaggle.com/datasets/johnsmith88/heart-disease-dataset

## 3. Evaluation 

> If we can reach about 95% accuracy with respect to the proof of concept then we'll pursue this project 

## 4. Features 

> **Data Features**
1. age - age in years
2. sex - (1 = male; 0 = female)
3. cp - chest pain type
    * 0: Typical angina: chest pain related decrease blood supply to the heart
    * 1: Atypical angina: chest pain not related to heart
    * 2: Non-anginal pain: typically esophageal spasms (non heart related)
    * 3: Asymptomatic: chest pain not showing signs of disease
4. trestbps - resting blood pressure (in mm Hg on admission to the hospital)
    * anything above 130-140 is typically cause for concern
5. chol - serum cholestoral in mg/dl
     * serum = LDL + HDL + .2 * triglycerides
     * above 200 is cause for concern
6. fbs - (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
    * '>126' mg/dL signals diabetes
7. restecg - resting electrocardiographic results
     * 0: Nothing to note
     * 1: ST-T Wave abnormality
           * can range from mild symptoms to severe problems
           * signals non-normal heart beat
     * 2: Possible or definite left ventricular hypertrophy
         *Enlarged heart's main pumping chamber
8. thalach - maximum heart rate achieved
9. exang - exercise induced angina (1 = yes; 0 = no)
10. oldpeak - ST depression induced by exercise relative to rest
    * looks at stress of heart during excercise
    * unhealthy heart will stress more
11. slope - the slope of the peak exercise ST segment
    * 0: Upsloping: better heart rate with excercise (uncommon)
    * 1: Flatsloping: minimal change (typical healthy heart)
    * 2: Downslopins: signs of unhealthy heart
12. ca - number of major vessels (0-3) colored by flourosopy
    * colored vessel means the doctor can see the blood passing through
    * the more blood movement the better (no clots)
13. thal - thalium stress result
    * 1,3: normal
    * 6: fixed defect: used to be defect but ok now
    * 7: reversable defect: no proper blood movement when excercising
14. target - have disease or not (1=yes, 0=no) (= the predicted attribute)

## Modelling 

```python
df.head()
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
## Splitting the data into train and test 
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

### After splitting the data, now lets evaluate the model by fitting and finding the accuracy of the model.
Let's use three models:-
1. Logistic regression 
2. K Nearest Neighbors
3. Random Forest classifier


```python
models={'Logistic Regression':LogisticRegression(),'KNN': KNeighborsClassifier(),'Random Forest':RandomForestClassifier()}
```

```python
def evaluation(models,X_train,y_train,X_test,y_test):
    """
    This function is designed to fit and calculate the accuracy of the model.
    models= The models we want to fit and score
    X_train=Training data(no labels)
    X_test=Testing data(no lables)
    y_train=Training data
    y_test=Testing data
    """
    model_scores={}
    for name,model in models.items():
        model.fit(X_train,y_train)
        model_scores[name]=model.score(X_test,y_test)
    
    return model_scores
```

```python
results=evaluation(models,X_train,y_train,X_test,y_test)
```

```python
results
```

```python
plt.bar(results.keys(),results.values());
plt.legend(['accuracy']);
```

## Hyperparameter tuning 

Lets try to tune the KNN model

```python
print(KNeighborsClassifier().get_params())
```

```python
knn=KNeighborsClassifier()
train_scores=[]
test_scores=[]
# Lets try to tune the n_neighbors
neighbors=np.arange(1,22)
for i in range(1,22):
    knn.set_params(n_neighbors=i)
    knn.fit(X_train,y_train)
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
print(max(train_scores),max(test_scores))
```

```python
print(neighbors)
```

```python
# seems like the accuracy hasn't increased notably 
plt.xticks(np.arange(1,22,1))
plt.plot(neighbors,train_scores)
plt.plot(neighbors,test_scores)
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.title('Hyperparameter tuning using n_neighbors')
plt.legend(['train scores','test scores'])
plt.figure(figsize=(15,6))
```

## Hyperparameter tuning using Randomizedsearch CV

We will tune the following models:-

1. LogisticRegression()
2. RandomForestClassifier()

```python
LogisticRegression().get_params()
```

```python
RandomForestClassifier().get_params()
```

```python
lr_grid= {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [100, 1000,2500, 5000]
    }

rf_grid={'n_estimators':np.arange(10,1000,50),'max_depth':[None,3,5,10],'min_samples_leaf':np.arange(1,20,2),'min_samples_split':np.arange(2,20,2),
        'max_features':np.arange(1,20,1)}
```

```python
# Now lets tune the logistic regression and find the results 
lr_model=RandomizedSearchCV(LogisticRegression(),lr_grid,cv=5,verbose=True,n_iter=20)
lr_model.fit(X_train,y_train)
```

```python
lr_model.best_params_
```

```python
lr_model.score(X_test,y_test)
```

```python
# Lets tune RandomForest and find the results 
rf_model=RandomizedSearchCV(RandomForestClassifier(),rf_grid,cv=5,verbose=True,n_iter=20)
rf_model.fit(X_train,y_train)
```

```python
rf_model.best_params_
```

```python
rf_model.score(X_test,y_test)
```

## Hyperparameter tuning using GridSearchCv 

So, we shall go with LogisticRegression as it has given the best accuracy so far

```python
grid={'penalty' : ['none'],
    'C' : [4.281332398719396],
    'solver' : ['liblinear','lbfgs'],
    'max_iter' : [100,1000]
    }

lr_gs=GridSearchCV(LogisticRegression(),grid,cv=5,verbose=True)
lr_gs.fit(X_train,y_train)
```

```python
lr_gs.best_params_
```

```python
lr_gs.score(X_test,y_test)
```
